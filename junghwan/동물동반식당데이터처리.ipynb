```python
import pandas as pd
from geojson import Point, Feature, FeatureCollection, dump
```

```python
data = pd.read_csv("../Data/정환/애견동반가능식당.csv")
```

```python
data
```

```python
coordinate_df = data[["name","lat","long"]]
```

```python
features = []
for i in range(0,len(coordinate_df)): 
    point = Point((coordinate_df.loc[i]["long"],coordinate_df.loc[i]["lat"]))
    features.append(Feature(geometry=point, properties={"name": coordinate_df.loc[i]["name"],"type":"동물동반식당"}))

with open('../Data/정환/동물동반식당.txt', 'w') as f:
    for item in features:
        f.write("%s\n" % item)
```

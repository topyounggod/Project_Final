```python
import pandas as pd
```

```python
data = pd.read_csv("../Data/정환/애견동반카페최종.csv", encoding="cp949")
```

```python
data
```

```python
coordinate_df = data[["사업장명","위도","경도"]]
```

```python
from geojson import Point, Feature, FeatureCollection, dump
```

```python
features = []
for i in range(0,len(coordinate_df)): 
    point = Point((coordinate_df.loc[i]['위도'],coordinate_df.loc[i]['경도']))
    features.append(Feature(geometry=point, properties={"name": coordinate_df.loc[i]["사업장명"],"type":"동물동반카페"}))

with open('../Data/정환/애견동반카페.txt', 'w') as f:
    for item in features:
        f.write("%s\n" % item)
```

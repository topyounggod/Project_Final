```python
def geturlNums(url):
    resultarr = []
    result = urlopen(url)
    html = result.read()
    soup = BeautifulSoup(html, 'html.parser')
    urls = soup.findAll("a",{"class":"moreBtn"})
    for a in urls:
        num = int(a["onclick"].split("'")[1])
        resultarr.append(num)
    
    return resultarr
```

```python
numarr = geturlNums(url)
```

```python
numarr
```

```python
def getTotalCount(url):
    result = urlopen(url)
    html = result.read()
    soup = BeautifulSoup(html, 'html.parser')
    tmp = soup.find("ul",{"class":"paging"}).findAll("li")[1].find("a")["href"]
    total = int(re.split('=|&',tmp)[1])
    
    return total
```

```python
startDate = "searchSDate=2021-10-06"
EndDate = "searchEDate=2021-10-06"
page = "https://www.animal.go.kr/front/awtis/protection/protectionList.do?"

cnt = getTotalCount(page + "&" + startDate + "&" + EndDate)

total_arr = []

for i in range(1,cnt//10+2):
    url = page + "&" + startDate + "&" + EndDate + "&page=" + str(i)
    total_arr = total_arr + geturlNums(url)
```

```python
import pandas as pd
from geojson import Point, Feature, FeatureCollection, dump
```

```python
data1 = pd.read_csv("../Data/정환/반려동물용품점.csv", encoding="cp949")
data2 = pd.read_csv("../Data/정환/애견동반가능식당.csv")
data3 = pd.read_csv("../Data/정환/애견동반카페최종.csv", encoding="cp949")

```

```python
data1
```

```python
coordinate_df = data1[["사업장명","주소","위도","경도"]]
```

```python
coordinate_df
```

```python
features = []
for i in range(0,len(coordinate_df)): 
    point = Point((coordinate_df.loc[i]["경도"], coordinate_df.loc[i]["위도"]))
    features.append(Feature(geometry=point, properties={"name": coordinate_df.loc[i]["사업장명"], "address":coordinate_df.loc[i]["주소"], "type":"동물용품판매점"}))

with open('../Data/정환/동물용품판매점.txt', 'w') as f:
    for item in features:
        f.write("%s\n" % item)
```

```python
coordinate = []
# 0.0005 = 500m
x = 37.584980
y = 127.031133
while x < 37.612181:
    y = 127.031133
    while y < 127.058599:
        coordinate.append((y,x))
        y += 0.0005
        y = round(y,6)
    coordinate.append((y,x))
    x += 0.0005
    x = round(x,6)
```

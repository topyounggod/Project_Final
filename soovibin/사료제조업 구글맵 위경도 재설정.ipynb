```python
import pandas as pd
import numpy as np
food = pd.read_csv("./data/서울특별시 사료제조업 인허가 정보.csv", index_col=0, encoding='CP949')
```

```python
food = food.reset_index()
food
```

```python
# 폐업 삭제
food = food[food['상세영업상태명']!='폐업']

import numpy as np
# 지번주소가 비어있는 곳 도로명주소로 채우기
food['지번주소'].replace(np.nan,'', inplace=True)
adds = food[['지번주소', '도로명주소']]
adds.index=range(0,len(adds))
# 지번주소 빈 칸인 곳에 도로명 주소 넣기!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
lists = []

for i in range(0, 270) :
    if adds['지번주소'][i] != '' :
        lists.append(adds['지번주소'][i])
    else:
        lists.append(adds['도로명주소'][i])
```

```python
food[food['지번주소'].isnull()]
```

```python
import googlemaps
import pandas as pd
places = food['지번주소']
my_key = ""
maps = googlemaps.Client(key='AIzaSyBe_EWzIxtwOQlPJAGqs7luMVboqNJAr0Y')  # my key값 입력
lat = []  #위도
lng = []  #경도

# 위치를 찾을 장소나 주소를 넣어준다.

i=0
for place in places:   
    i = i + 1
    try:
        print("%d번 인덱스에서 %s의 위치를 찾고있습니다"%(i, place))
        geo_location = maps.geocode(place)[0].get('geometry')
        lat.append(geo_location['location']['lat'])
        lng.append(geo_location['location']['lng'])
        

    except:
        lat.append('')
        lng.append('')
        print("%d번 인덱스 위치를 찾는데 실패했습니다."%(i))


# 데이터프레임만들어 출력하기
df = pd.DataFrame({'위도':lat, '경도':lng})
df
```

```python
result = pd.concat([df,food], axis=1, sort=False)
result
```

```python
result = result.drop(['좌표정보(X)','좌표정보(Y)'], axis='columns')
result
```

```python
# csv 저장
result.to_csv('./data/pet_foodmake_raw.csv')
```

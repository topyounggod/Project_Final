```python
import pandas as pd
a = pd.read_csv("./data/animal_infra.csv", index_col=0)
```

```python
del a['Unnamed: 0.1']
a
```

```python
# pd.set_option('display.max_rows', None) # 전체행 출력해주는 설정명령어
pd.reset_option('display.max_rows', None) # 원래 설정으로
```

```python
a = a[~a['type'].str.contains('동물보호센터')]
a = a[~a['type'].str.contains('동물장묘업')]
a = a[~a['type'].str.contains('동물용의료용구판매업')]
a
```

```python
hos = a[a['type'] == '동물병원']
petpharmacy = a[a['type'] == '동물약국업']
petplay = a[a['type'] == '동물놀이터']
petpark = a[a['type'] == '동물공원']
동물동반식당 = a[a['type'] == '동물동반식당']
동물동반카페 = a[a['type'] == '동물동반카페']
petbeauty = a[a['type'] == '동물미용업']
pettaxi = a[a['type'] == '동물운송업']
반려동물용품점 = a[a['type'] == '반려동물용품점']
petsell = a[a['type'] == '동물판매업']
feedmake = a[a['type'] == '사료제조업']
위탁관리 = a[a['type'] == '위탁관리업']
```

```python
hos['type_id'] = '1'
petpharmacy['type_id'] = '2'
petplay['type_id'] = '3'
petpark['type_id'] = '4'
동물동반식당['type_id'] = '5'
동물동반카페['type_id'] = '6'
petbeauty['type_id'] = '7'
pettaxi['type_id'] = '8'
반려동물용품점['type_id'] = '9'
petsell['type_id'] = '10'
feedmake['type_id'] = '11'
위탁관리['type_id'] = '12'
```

```python
result = pd.concat([hos, petpharmacy, petplay, petpark, 동물동반식당,동물동반카페, petbeauty, pettaxi, 
                    반려동물용품점, petsell, feedmake, 위탁관리])
result
```

```python
del result['type']
result
```

```python
# 저장
result.to_csv('./data/typeid_fin.csv')
```

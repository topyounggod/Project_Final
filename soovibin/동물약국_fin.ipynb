```python
import pandas as pd
import numpy as np
```

```python
a = pd.read_csv('./data/서울특별시 동물약국 인허가 정보(utf-8).csv',encoding = 'utf8')
a = a[a['상세영업상태명'] == '정상']
a['상세영업상태명'].unique()
```

```python
# 결측치 빈칸으루
a['지번주소'].replace(np.nan,'', inplace=True)

adds = a[['지번주소', '도로명주소']]
```

```python
adds['지번주소'].isnull().sum()
```

```python
adds['도로명주소'].isnull().sum()
```

```python
# 지번주소 빈 칸인 곳에 도로명 주소 넣기!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
lists = []

for i in range(0, len(adds)) :
    if adds['지번주소'][i] != '' :
        lists.append(adds['지번주소'][i])
    else:
        lists.append(adds['도로명주소'][i]) 
```

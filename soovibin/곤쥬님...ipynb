```python

import numpy as np

import pandas as pd



import matplotlib.pyplot as plt

import seaborn as sns

plt.rcParams['figure.figsize'] = [10, 8]



```

```python
df = pd.read_csv("gongju.csv")
```

```python
del df['Unnamed: 0']
```

```python
del df['address.1']
del df['type.1']
```

```python
df
```

```python
df.dropna(inplace=True)
```

```python
df1 = df[df.address.str.contains('서울')]
```

```python
df1.reset_index(drop=True,inplace=True)
```

```python
pdf1 = pd.pivot_table(df1,
                     index='시군구',
                     columns='type',
                     values='address',
                     aggfunc='count')
pdf1
```

```python
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname = 'C:\windows/fonts/malgun.ttf').get_name()
rc('font',family = font_name)
```

```python
plt.rcParams['axes.unicode_minus'] = False
ax = sns.heatmap(pdf1, cmap='RdYlGn_r')
# sns.heatmap(df, annot=True, fmt='d')
plt.title('구별 타입 개수', fontsize=20)


plt.show() 

## 숫자 넣을지 말지는 내일 얘기하자구~
```

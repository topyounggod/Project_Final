```python
import pandas as pd
import numpy as np

animal_funeral = pd.read_csv('./data/동물장묘업최최종(UTF-8).csv',encoding = 'utf-8')
```

```python
animal_funeral
```

```python
import googlemaps
import pandas as pd
places = animal_funeral['업체명']
my_key = ""
maps = googlemaps.Client(key=my_key)  # my key값 입력
lat = []  #위도
lng = []  #경도

# 위치를 찾을 장소나 주소를 넣어준다.

i=0
for place in places:   
    i = i + 1
    try:
        print("%d번 인덱스에서 %s의 위치를 찾고있습니다"%(i, place))
        geo_location = maps.geocode(place)[0].get('geometry')
        lat.append(geo_location['location']['lat'])
        lng.append(geo_location['location']['lng'])
        

    except:
        lat.append('')
        lng.append('')
        print("%d번 인덱스 위치를 찾는데 실패했습니다."%(i))


# 데이터프레임만들어 출력하기
df = pd.DataFrame({'위도':lat, '경도':lng})
df
```

```python
df_fin = pd.concat([animal_funeral,df],axis = 1)

df_fin.to_csv('동물장묘업경도(euc-kr).csv',encoding='euc-kr')
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
import numpy as np
```

```python
url_init = "https://www.animal.go.kr/front/awtis/protection/protectionDtl.do?boardId=&desertionNo=447505202100858"
result = urlopen(url_init)
html = result.read()
soup = BeautifulSoup(html, 'html.parser')
```

```python
# 주피터에서 데이터프레임 변화시키려고했지만
# 코드가 너무 복잡해져서
# engineering 분들이 html에서 테이블 요소들 바로바로 찾아서 df로 만드는 과정을 거침.


table = soup.select('table')
cols = soup.find("table").findAll("th")
soup.find("table").findAll("td")
soup.find("table").findAll("td")[0].text.strip()
dic = {}
for i in range(len(cols)):
    dic[soup.find("table").findAll("th")[i].text] = soup.find("table").findAll("td")[i].text.strip()
animals = pd.DataFrame.from_dict([dic])
animals.drop([0], inplace=True)
animals
```

```python
import pandas as pd
import numpy as np
 
animal_equip = pd.read_csv('D:/OneDrive/DataScience/프로젝트(파이널)/Data/서울열린데이터광장/서울특별시 동물용의료용구판매업 인허가 정보.csv',encoding = 'euc-kr')
```

```python
animal_equip
```

```python
animal_equip= animal_equip[animal_equip['상세영업상태명']!='폐업']
```

```python
animal_equip
```

```python
import googlemaps
import pandas as pd
places = animal_equip['도로명주소']
my_key = ""
maps = googlemaps.Client(key=my_key)  # my key값 입력
lat = []  #위도
lng = []  #경도

# 위치를 찾을 장소나 주소를 넣어준다.

i=0
for place in places:   
    i = i + 1
    try:
        print("%d번 인덱스에서 %s의 위치를 찾고있습니다"%(i, place))
        geo_location = maps.geocode(place)[0].get('geometry')
        lat.append(geo_location['location']['lat'])
        lng.append(geo_location['location']['lng'])
        

    except:
        lat.append('')
        lng.append('')
        print("%d번 인덱스 위치를 찾는데 실패했습니다."%(i))


# 데이터프레임만들어 출력하기
df = pd.DataFrame({'위도':lat, '경도':lng})
df
```

```python
df_fin = pd.concat([animal_equip,df],axis = 1)

df_fin.to_csv('동물의료용구판매업(utf-8).csv',encoding='utf-8')
```

```python
import pandas as pd
df = pd.read_csv('./data/서울시주요공원현황.csv',encoding = 'euc-kr')
```

```python
df.isnull().sum()
```

```python
df['type'] = '동물공원'
```

```python
df
```

```python
df = df[['공원명','위도','경도','공원주소','type']]
```

```python
df = df.rename(columns={'공원명':'name','위도':'lat','경도':'lng','공원주소':'address'})
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.to_csv('동물공원(utf-8)최종_영훈.csv',encoding='utf-8')
```

```python
import geojson
from geojson import Point, Feature, FeatureCollection, dump
```

```python
features = []
for i in range(0,len(df)): 
    #print(coordinate_df.loc[i])
    #print(coordinate_df.loc[i]['위도'])
    #print(coordinate_df.loc[i]['경도'])
    point = Point((df.loc[i]['lng'],df.loc[i]['lat']))
    features.append(Feature(geometry=point, properties={"name": df.loc[i]['name'],"address":df.loc[i]['address'],"type":df.loc[i]['type']}))

with open('동물공원.txt', 'w') as f:
    for item in features:
        f.write("%s\n" % item)
```
